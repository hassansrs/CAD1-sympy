```python
n,p=symbols('n,p',positive=True)
a_n=(2/p)*(integrate(t*cos(2*pi*n*t/p),t,(t,-p/2,p/2)))
b_n=(2/p)*(integrate(t*sin(2*pi*n*t/p),t,(t,-p/2,p/2)))
a_n.trigsimp(),b_n.trigsimp()
```
